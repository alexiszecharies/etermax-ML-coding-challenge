```python
df['device_family'] = df['device_family'].str.lower()

def map_dispositivo(dispo):#solo me quedo con dispositivos comunes
    if type(dispo) == str:
        if 'xiaomi' in dispo:
            return 'xiaomi phone'
        elif 'motorola' in dispo:
            return 'motorola phone'
        elif 'oppo' in dispo:
            return 'oppo phone'
        elif 'ipad' in dispo:
            return 'apple ipad'
        elif 'huawei' in dispo:
            return 'huawei phone'
        elif 'samsung galaxy' in dispo:
            return 'samsung galaxy phone'
        elif 'samsung samsung' in dispo:
            # dejo esto como otra categoria porque los galaxy suelen ser mas premium
            return 'samsung no galaxy phone'
        elif 'lg' in dispo:
            return 'lg phone'
        elif 'sony' in dispo:
            return 'sony phone'
        elif 'lenovo' in dispo:
            return 'lenovo device'
        elif 'bq' in dispo:
            return 'bq phone'
        elif 'tcl' in dispo:
            return 'tcl device'
        elif 'realme' in dispo:
            return 'realme device'
        elif 'asus' in dispo:
            return 'asus device'
        elif 'oneplus' in dispo:
            return 'oneplus device'
        elif dispo == 'apple iphone':
            return dispo
        else:
            return 'otros'
    
df['device_family'] = df['device_family'].apply(lambda x : map_dispositivo(x))
```

#### Event 1 es muy paricular, parece tener mas del 99% de ceros. Para event 2 y 3 tomemos el criterio de que todo lo que sea mas grande que el percentil 99 queda topeado con ese valor

```python
def setear_max(df, col, nuevo_max):
    df[col] = df.apply(lambda x: nuevo_max if x[col] > nuevo_max else x[col], axis = 1)
    return df
```

```python
percentil_99 = df[['event_1', 'event_2','event_3', 'revenue']].describe(percentiles=np.arange(0.01,1.0,0.01)).loc["99%", :]

for col in ['event_2','event_3']:
    df = setear_max(df, col, percentil_99[col])
```

#### Otra variable que hay que manejar es la version del os porque tiene un monton de categorias. Probemos solo dejando una cifra despues del punto a ver como se distribuye. No deberia haber cambios enormes de una 10.0.1 a una 10.0.3

```python
def solo_una_cifra_despues_del_punto(version_os):
    version_split = version_os.split('.')
    if len(version_split) > 2:
        version_os = version_split[0] + '.' + version_split[1]
    return version_os
df['os_version'] = df['os_version'].apply(lambda x: solo_una_cifra_despues_del_punto(x))
```

```python
os_version_frec = (df['os_version'].value_counts())/df.shape[0]
```

```python
os_version_frec
```

```python
os_version_con_menor_frec = os_version_frec[os_version_frec<=0.01]# a los que aparecen menos de un 1% los agrupo
df.loc[df["os_version"].isin(os_version_con_menor_frec.index.tolist()), 'os_version']= "otra_version"
```

#### Voy a hacer un hibrido donde las ciudades mas frecuentas quedan con el dato de la ciudad y cuando no se cumpla eso, voy  a mirar si el pais es uno de los que tienen una muestra considerable y en caso de que no sea asi los agruparemos en una categoria "otros paises"

```python
country_region_frec = (df['country_region'].value_counts())/df.shape[0]

country_region_con_menor_frec = country_region_frec[country_region_frec<=0.01]# a los que aparecen menos de un 1% los agrupo
df.loc[df["country_region"].isin(country_region_con_menor_frec.index.tolist()), 'country_region']= "otras_ciudades"
```

```python
df['country_region'].value_counts()
```

```python
ciudades_populares = country_region_frec[~(country_region_frec<=0.01)].index
ciudades_populares
```

```python
country_frec = (df['country'].value_counts())/df.shape[0]

country_con_menor_frec = country_frec[country_frec<=0.01]# a los que aparecen menos de un 1% los agrupo
df.loc[df["country"].isin(country_con_menor_frec.index.tolist()), 'country']= "otros_paises"
```

```python
df['country'].value_counts()
```

```python
paises_populares = country_frec[~(country_frec<=0.01)].index
paises_populares
```

#### Creo una nueva variable place que sea un hibrido entre ciudades y paises grandes y hago una categoria nueva que se llame lugar_poco_frecuente con los datos que no caigan en esos paises y ciudades

```python
def transformar_ciudad_pais(pais, ciudad):
    if ciudad in ciudades_populares:
        return ciudad
    elif pais in paises_populares:
        return pais
    else:
        return "lugar_poco_frecuente"
df['place'] = df.apply(lambda x: transformar_ciudad_pais(x['country'], x['country_region']), axis = 1)
df['place'].value_counts()
```

```python
df.columns
```

```python
columnas_modelo = ['source', 'place', 'platform', 'device_family', 'os_version', 'event_1', 'event_2', 'event_3', 'revenue']
df = df[columnas_modelo]
```

#### Pasamos a dummies las var categoricas

```python
col_categoricas = ['source', 'place', 'platform', 'device_family', 'os_version']
for col_cat in col_categoricas:
    one_hot_data = pd.get_dummies(df[col_cat])
    df = pd.concat([one_hot_data, df[[col for col in df.columns if col != col_cat]]], axis=1)
```

```python
df
```

```python
print(df.columns)
```

```python
df.columns = [x.lower() for x in df.columns]
```

```python
X = df.drop(columns=['revenue'])
Y = df['revenue']
```

#### Divido en train y test

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state=4)
```

#### Estandarizo mi muestra para mejores resultados

```python
scaler_x = MinMaxScaler()
X_train_std = scaler_x.fit_transform(X_train)
X_test_std = scaler_x.transform(X_test)
```

### Hagamos una simple regresion lineal

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train_std, y_train)
```

```python
print("score en train R2:", lr.score(X_train_std, y_train))
```

```python
n_scores_lr = cross_val_score(lr, X_test_std, y_test, scoring='r2', cv=4)
print('score en test','R2: %.3f (%.3f)' % (np.mean(n_scores_lr), np.std(n_scores_lr)))
```

### Intentemos con algoritmos mas complejos y con grid search para optimizar puntaje asi, pero sin mucho max depth teniendo en cuenta que la respuesta debe ser rapida al productivizar. Usamos cv para lograr resultados mas consistentes

```python
GBR = GradientBoostingRegressor()

parameters = {'learning_rate': [0.01,0.02,0.03,0.10],
                  'subsample'    : [0.9, 0.5, 0.2, 0.1],
                  'n_estimators' : [100,500,1000, 1500],
                  'max_depth'    : [3,4,5,6]
                 }
grid_GBR = GridSearchCV(estimator=GBR, param_grid = parameters, cv = 3, n_jobs=-1)
grid_GBR.fit(X_train_std, y_train)
```

```python
print(" Results from Grid Search " )
print("\n The best estimator across ALL searched params:\n",grid_GBR.best_estimator_)
print("\n The best score across ALL searched params:\n",grid_GBR.best_score_)
print("\n The best parameters across ALL searched params:\n",grid_GBR.best_params_)
```

```python
n_scores_GBR = cross_val_score(grid_GBR.best_estimator_, X_test_std, y_test, scoring='r2', cv=4)
print('R2: %.3f (%.3f)' % (np.mean(n_scores_GBR), np.std(n_scores_GBR)))
```

```python
xgb = XGBRegressor()
parameters = {'objective':['reg:linear'],
              'learning_rate': [0.01,0.02,0.07,0.10],
              'max_depth': [5, 6, 7],
              'n_estimators': [100,500,1000, 1500]}

xgb_grid = GridSearchCV(xgb, parameters, cv = 3, n_jobs = -1)

xgb_grid.fit(X_train_std, y_train)
```

```python
print(" Results from Grid Search " )
print("\n The best estimator across ALL searched params:\n",xgb_grid.best_estimator_)
print("\n The best score across ALL searched params:\n",xgb_grid.best_score_)
print("\n The best parameters across ALL searched params:\n",xgb_grid.best_params_)
```

```python
n_scores_xgb = cross_val_score(xgb_grid.best_estimator_, X_test_std, y_test, scoring='r2', cv=3)
print('R2: %.3f (%.3f)' % (np.mean(n_scores_xgb), np.std(n_scores_xgb)))
```

#### El mejor modelo en test fue el gradient boosting tree. Lo guardamos junto con sus parametros optimos en la carpeta model junto con otras variables necesarias para el servidor flask

```python
joblib.dump(grid_GBR.best_estimator_, "model/modelo_GBR.pkl")
joblib.dump(scaler_x, "model/standarizer.pkl")
joblib.dump(moda_por_pais, "model/moda_por_pais.pkl")
joblib.dump(percentil_99, "model/percentil_99.pkl")
joblib.dump(os_version_con_menor_frec, "model/os_version_con_menor_frec.pkl")
joblib.dump(country_region_con_menor_frec, "model/country_region_con_menor_frec.pkl")
```
